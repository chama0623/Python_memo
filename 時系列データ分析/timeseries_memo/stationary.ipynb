## check statinary

### example : daily-total-female-births time series

```python
# histgram
series = pd.read_csv('daily-total-female-births.csv', header=0, index_col=0)
series.hist()
plt.show()
```

```python
# split time series into tow contiguous sequences
# calculate the mean and variance
X = series.values
split = round(len(X) / 2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

### example : airline-passengers time series

```python
series = pd.read_csv('airline-passengers.csv', header=0, index_col=0)
series.hist()
plt.show()
```

```python
## log translator
X = series.values
X = np.log(X)
plt.hist(X)
plt.show()
plt.plot(X)
plt.show()
```

```python
# calculate the mean and variance
split = round(len(X) / 2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

## ADF test

### example : daily-total-female-births time series

```python
from statsmodels.tsa.stattools import adfuller
```

```python
series = pd.read_csv('daily-total-female-births.csv', header=0, index_col=0)
X = series.values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

### example : airline-passengers time series

```python
series = pd.read_csv('airline-passengers.csv', header=0, index_col=0)
X = series.values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```
