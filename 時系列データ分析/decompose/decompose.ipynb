```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("20201114_tokyo_covid19_patients.csv")
df = df.drop(columns=["全国地方公共団体コード","市区町村名","発症_年月日","確定_年月日",'患者_年代', '患者_性別', '患者_職業', '患者_状態', '患者_症状', '患者_渡航歴の有無フラグ',
       '患者_接触歴の有無フラグ', '備考', '退院済フラグ'],axis=1)
df['公表_年月日'] = pd.to_datetime(df['公表_年月日'], format="%Y-%m-%d")
df.head()
```

```python
patients =  pd.DataFrame(df['公表_年月日'].value_counts())
patients.reset_index(inplace=True)
patients = patients.sort_values("index").reset_index()
patients = patients.drop(columns=["level_0"],axis=1)
patients = patients.rename(columns={"index":"date","公表_年月日":"陽性者数"})
patients
```

```python
plt.figure(figsize=(20,5),facecolor="white")
plt.plot(patients["date"],patients["陽性者数"])
```

```python
import statsmodels.api as sm

res = sm.tsa.seasonal_decompose(patients["陽性者数"],period=31)
trend = res.trend # トレンドデータ
seasonality = res.seasonal # 季節性データ
residual = res.resid # 残差データ

fig = plt.figure(figsize=(20,10),facecolor="white")
fig.suptitle("東京都covid-19 新規陽性者 時系列データの分解",fontsize=20)
plt.subplot(4,1,1)
plt.plot(patients["date"],patients["陽性者数"])
plt.ylabel('Original',fontsize=17)
plt.subplot(4,1,2) 
plt.plot(trend)
plt.ylabel('trend',fontsize=17)
plt.subplot(4,1,3)
plt.plot(seasonality)
plt.ylabel('seasonality',fontsize=17)
plt.subplot(4,1,4) 
plt.plot(residual)
plt.ylabel('residual',fontsize=17)

plt.savefig("20201114_tokyo_covid19_decompose.png")
plt.show()
```
