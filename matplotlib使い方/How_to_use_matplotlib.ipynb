### タイトル,軸ラベルをつける

````python
plt.figure(facecolor="white") # ダークモードを使ってる人は実行してね
plt.scatter(df["x"],df["y"]) # 点だけ表示したいとき
plt.plot(df["x"],df["y"]) # 折れ線グラフで表示したいとき

# タイトル,軸ラベルをつけるコード
# plt.title("タイトルです") # タイトルをつける
# plt.xlabel("x軸のラベルです") # x軸ラベルをつける
# plt.ylabel("y軸のラベルです") # y軸ラベルをつける

# フォントサイズを設定したい場合
plt.title("タイトルです",fontsize=20)
plt.xlabel("x軸のラベルです",fontsize=15) 
plt.ylabel("y軸のラベルです",fontsize=10)

plt.show()
````

### 目盛りを設定する

````python
# 元のグラフ
plt.figure(facecolor="white") # ダークモードを使ってる人は実行してね
plt.scatter(df["x"],df["y"]) # 点だけ表示したいとき
plt.plot(df["x"],df["y"]) # 折れ線グラフで表示したいとき
plt.show()
````

````python
# 目盛りを設定したグラフ
plt.figure(facecolor="white") # ダークモードを使ってる人は実行してね
plt.scatter(df["x"],df["y"]) # 点だけ表示したいとき
plt.plot(df["x"],df["y"]) # 折れ線グラフで表示したいとき
plt.xlim(0,3) # x軸の目盛りを設定(start,end)
plt.ylim(0,25) # y軸の目盛りを設定(start,end)
plt.show()
````

### グリッドを表示する

### ラベルをつける

````python
plt.figure(facecolor="white") # ダークモードを使ってる人は実行してね
plt.scatter(df["x"],df["y"],label="東京") # 点だけ表示したいとき
plt.scatter(df["x"],df["y"]+10,label="大阪") # 折れ線グラフで表示したいとき
plt.legend() # ラベルを表示
# plt.legend(fontsize=15) # ラベルサイズも変えられるよ
plt.show()
````

### 点の形,色を変える

### その他細かい体裁を整える

````python
plt.figure(facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.scatter(df["x"],df["y"],label="東京") 
plt.scatter(df["x"],df["y"]+10,label="大阪") 
plt.legend()
plt.show()
````

## グラフを並べて表示する
subplot(縦表示数,横表示数,インデックス)で複数描画が可能

````python
# 縦に並べる場合
plt.figure(facecolor="white")
plt.subplot(2,1,1) # グラフ1を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(2,1,1)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.subplot(2,1,2) # グラフ2を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(2,1,2)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.show()
````

````python
# 横に並べる場合
plt.figure(figsize=(10,5),facecolor="white") # figsizeでグラフのサイズを変更
plt.subplot(1,2,1) # グラフ1を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(1,2,1)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.subplot(1,2,2) # グラフ2を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(1,2,2)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.show()
````

````python
# 複数表示
plt.figure(figsize=(10,5),facecolor="white") # figsizeでグラフのサイズを変更
plt.subplot(2,2,1) # グラフ1を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(2,2,1)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.subplot(2,2,2) # グラフ2を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(2,2,2)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.subplot(2,2,3) # グラフ3を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(2,2,3)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.subplot(2,2,4) # グラフ4を描画
plt.scatter(df["x"],df["y"]) 
plt.text(4, 10, "(2,2,4)",ha='center',fontsize=14) # どのグラフか表示(実際はいらないから消してね)
plt.show()
````

## 散布図に最小二乗法直線を引く
最小二乗法の計算を行うためにscikit-learnというライブラリを用います.インストールは
``pip install sklearn``で可能です

````python
# プロットするデータを入力してね
x_data = df["x"]
y_data = df["y"]

###############################

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression() # インスタンス化

# x,yを行列に変換
x = x_data.values.reshape(-1,1)
y = y_data.values.reshape(-1,1)
# fitting
lin_reg.fit(x,y)
x_pred = np.linspace(x_data.min()-2,x_data.max()+2,len(x_data)+4).reshape(-1,1)
y_pred = lin_reg.predict(x_pred)

plt.figure(facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )

plt.scatter(x_data,y_data,label="測定結果")
plt.plot(x_pred,y_pred,label="最小二乗曲線")
plt.xlim(x_data.min()-1,x_data.max()+1)
plt.ylim(0,60) # 調整してね
plt.legend()
plt.show()
````
