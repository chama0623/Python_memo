# Ridge regresion

## linear

### make data

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
m=50
X = 6*np.random.rand(m,1)-3 #ranedom.rand(m,n)   m*n行列で0以上1未満の乱数を生成する
y=0.9*X+2+np.random.randn(m,1) #random.randn(m,n)   m*n行列で平均0,分散0の乱数を生成する
```

```python
plt.figure(facecolor="white")
plt.scatter(X,y)
plt.show()
```

### modeling

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1) # インスタンス化
ridge_reg.fit(X,y)
ridge_reg.intercept_,ridge_reg.coef_
```

```python
plt.figure(facecolor="white")
plt.scatter(X,y)
plt.plot(X,ridge_reg.predict(X),"r-")
plt.show()
```

## Non-linear

## make data

```python
m=50
degree = 4
x = 8*np.random.rand(m,1)-1.5 #ranedom.rand(m,n)   m*n行列で0以上1未満の乱数を生成する
X=np.zeros((m,degree)) #データサイズ*次元数の配列を生成
X[:, 0] = x.T #0列目にzを代入
for i in range(degree-1): #各次元におけるxの値を計算
    X[:,i+1] = X[:,i]*X[:,0]
y = np.sin(X[:,0]).reshape(-1,1)+ np.random.randn(m,1)/2.70 + 8.5  #yを生成
```

```python
plt.figure(facecolor="white")
plt.scatter(X.T[0],y)
plt.show()
```

### modeling

```python
ridge_reg = Ridge(alpha=1)
ridge_reg.fit(X,y)
ridge_reg.intercept_,ridge_reg.coef_
```

```python
plot_num = 200
x_test = np.array([np.arange(-10,10,0.1)]).reshape(plot_num,1) #グラフのためのnumpy配列を生成

X_test=np.zeros((plot_num,degree)) #データサイズ*次元数の配列を生成
X_test[:, 0] = x_test.T #0列目にzを代入
for i in range(degree-1): #各次元におけるxの値を計算
    X_test[:,i+1] = X_test[:,i]*X_test[:,0]
```

```python
plt.figure(facecolor="white")
plt.scatter(x,y)
plt.plot(x_test,ridge_reg.predict(X_test),"r-")
plt.xlim(-2,8)
plt.ylim(6.5,10.5)
plt.show()
```

### change hyperparameter

```python
ridge_reg1 = Ridge(alpha=1)
ridge_reg2 = Ridge(alpha=10)
ridge_reg3 = Ridge(alpha=100)
ridge_reg1.fit(X,y)
ridge_reg2.fit(X,y)
ridge_reg3.fit(X,y)

plt.scatter(x,y)
plt.plot(x_test,ridge_reg1.predict(X_test),"r-",label="a=    1")
plt.plot(x_test,ridge_reg2.predict(X_test),"g-",label="a=  10")
plt.plot(x_test,ridge_reg3.predict(X_test),"m-",label="a=100")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,8)
plt.ylim(6.5,10.5)
plt.grid()
plt.legend()
plt.show()
```
