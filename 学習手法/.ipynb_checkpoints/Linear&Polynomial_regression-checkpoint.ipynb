## read library

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Make data

```python
N=30
X=2*np.random.rand(N,1)
y=4+3*X+np.random.randn(N,1)
```

```python
plt.figure(facecolor='white')
plt.scatter(X,y,label='point')
plt.legend()
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression() # インスタンス化
model.fit(X,y) # 学習を実行
model.intercept_[0],model.coef_[0][0] # 切片(intercept),傾き(coef)を表示
```

```python
point = np.arange(X.min(),X.max(),0.1).reshape(-1,1) # arangeを生成
predict = model.predict(point)

plt.figure(facecolor='white')
plt.scatter(X,y,label='point')
plt.plot(point,predict,color='red',label='regression line')
plt.legend()
```
