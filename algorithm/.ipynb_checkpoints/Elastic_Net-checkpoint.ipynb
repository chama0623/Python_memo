# read library

```python
import numpy as np
from matplotlib import pyplot as plt
```

# linear

## make data

```python
N=30
X=2*np.random.rand(N,1)
y=4+3*X+np.random.randn(N,1)
```

```python
plt.figure(facecolor='white')
plt.scatter(X,y,label='point')
plt.legend()
```

## modeling

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1,l1_ratio=0.5) # l1_ratio means r
elastic_net.fit(X,y)
elastic_net.intercept_,elastic_net.coef_
```

```python
plt.figure(facecolor="white")
plt.scatter(X,y)
plt.plot(X,elastic_net.predict(X),"r-")
plt.show()
```

```python
elastic_net_reg1 = ElasticNet(alpha=0.1,l1_ratio=0.5)
elastic_net_reg2 = ElasticNet(alpha=1,l1_ratio=0.5)
elastic_net_reg3 = ElasticNet(alpha=0.1,l1_ratio=1)
elastic_net_reg1.fit(X,y)
elastic_net_reg2.fit(X,y)
elastic_net_reg3.fit(X,y)

plt.figure(facecolor="white")
plt.scatter(X,y)
plt.plot(X,elastic_net_reg1.predict(X),"r-",label="alpha=0.1,r=0.5")
plt.plot(X,elastic_net_reg2.predict(X),"g-",label="alpha=  1,r=0.5")
plt.plot(X,elastic_net_reg3.predict(X),"m-",label="alpha=0.1,r=  1")
plt.xlabel("x")
plt.ylabel("y")
#plt.xlim(-2,8)
#plt.ylim(6.5,10.5)
plt.grid()
plt.legend()
plt.show()
```

## Non linear

```python
m=50
degree = 4
x = 8*np.random.rand(m,1)-1.5 #ranedom.rand(m,n)   m*n行列で0以上1未満の乱数を生成する
X=np.zeros((m,degree)) #データサイズ*次元数の配列を生成
X[:, 0] = x.T #0列目にzを代入
for i in range(degree-1): #各次元におけるxの値を計算
    X[:,i+1] = X[:,i]*X[:,0]
y = np.sin(X[:,0]).reshape(-1,1)+ np.random.randn(m,1)/2.70 + 8.5  #yを生成
```

```python
plt.figure(facecolor="white")
plt.scatter(X.T[0],y)
plt.show()
```

```python
elastic_net = ElasticNet(alpha=0.1,l1_ratio=0.5) # l1_ratio means r
elastic_net.fit(X,y)
elastic_net.intercept_,elastic_net.coef_
```

```python
plot_num = 200
x_test = np.array([np.arange(-10,10,0.1)]).reshape(plot_num,1) #グラフのためのnumpy配列を生成

X_test=np.zeros((plot_num,degree)) #データサイズ*次元数の配列を生成
X_test[:, 0] = x_test.T #0列目にzを代入
for i in range(degree-1): #各次元におけるxの値を計算
    X_test[:,i+1] = X_test[:,i]*X_test[:,0]
```

```python
plt.figure(facecolor="white")
plt.scatter(x,y)
plt.plot(x_test,elastic_net.predict(X_test),"r-")
plt.xlim(-2,8)
plt.ylim(6.5,10.5)
plt.show()
```

```python
elastic_net1 = ElasticNet(alpha=0.1,l1_ratio=0.5)
elastic_net2 = ElasticNet(alpha=1,l1_ratio=0.5)
elastic_net3 = ElasticNet(alpha=0.1,l1_ratio=1)
elastic_net1.fit(X,y)
elastic_net2.fit(X,y)
elastic_net3.fit(X,y)

plt.figure(facecolor="white")
plt.scatter(x,y)
plt.plot(x_test,elastic_net1.predict(X_test),"r-",label="alpha=0.1,r=0.5")
plt.plot(x_test,elastic_net2.predict(X_test),"g-",label="alpha=  1,r=0.5")
plt.plot(x_test,elastic_net3.predict(X_test),"m-",label="alpha=0.1,r=  1")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,8)
plt.ylim(6.5,10.5)
plt.grid()
plt.legend()
plt.show()
```
